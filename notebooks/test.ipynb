```python
import sys, os
from pathlib import Path

# ==========================================================
# 🔍 1️⃣  DETECT PROJECT ROOT SAFELY
# ==========================================================
ROOT = Path.cwd()

# if notebook is inside /notebooks → move 1 level up
if ROOT.name.lower() in ["notebook", "notebooks"]:
    ROOT = ROOT.parent

# add project + ecoml package to python path
sys.path.insert(0, str(ROOT))
sys.path.insert(0, str(ROOT / "ecoml"))

print("📁 Project Root =", ROOT)

# ==========================================================
# 🔌 2️⃣  IMPORT ECO-ML COMPONENTS
# ==========================================================
from ecoml import EcoTracker, CellHook, RecommendationEngine, GeminiAdvisor

# ==========================================================
# 🌱 3️⃣  START TRACKER
# ==========================================================
tracker = EcoTracker(
    log_path=str(ROOT / "data" / "emissions_log.csv"),
    notebook_name="TEST_NOTEBOOK"
)

# ==========================================================
# 🤖 4️⃣  GEMINI + RECOMMENDER
# ==========================================================
gem = GeminiAdvisor()
rec = RecommendationEngine(tracker=tracker, gemini=gem)

# ==========================================================
# 🧩 5️⃣  ACTIVATE UI HOOK
# ==========================================================
CellHook(tracker, rec).register()

print("\n🔥 EcoML READY")
print("🤖 Gemini:", "ON ✓" if gem.enabled else "OFF ❌")
```
